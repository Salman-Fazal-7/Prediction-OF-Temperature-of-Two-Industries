# Polynomial Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Dataset

```python
dataset= pd.read_csv('/content/annual_temp.csv')
dataset
```

# **Separate** **the** **data** **into** **two** **datasets**,**one** **for** **each** **Industry**

# INDUSTRY1

```python
Industry1 = dataset[dataset["Source"].str.contains("GCAG")]
Industry1
```

# INDUSTRY2

```python
Industry2 = dataset[dataset["Source"].str.contains("GISTEMP")]
Industry2
```

# Data Pre-processing

# INDUSTRY

```python
x1 = Industry1.iloc[:,1:2].values
y1 = Industry1.iloc[:,-1].values
```

```python
x1
```

```python
y1
```

# INDUSTRY2

```python
x2 = Industry2.iloc[:,1:2].values
y2 = Industry2.iloc[:,-1].values
```

```python
x2
```

```python
y2
```

## Training the Linear Regression model on the whole dataset

# INDUSTRY1

```python
from sklearn.linear_model import LinearRegression
lin_reg1 = LinearRegression()
lin_reg1.fit(x1, y1)
```

# INDUSTRY2

```python
from sklearn.linear_model import LinearRegression
lin_reg2 = LinearRegression()
lin_reg2.fit(x2, y2)
```

## Visualising the Linear Regression results

# INDUSTRY1

```python
plt.scatter(x1, y1, color = 'green')
plt.plot(x1, lin_reg1.predict(x1), color = 'black')
plt.title('Temperature of GCAG (Linear Regression)')
plt.xlabel('years')
plt.ylabel('Means')
plt.show()
```

# INDUSTRY2

```python
plt.scatter(x2, y2, color = 'green')
plt.plot(x2, lin_reg2.predict(x2), color = 'BLACK')
plt.title('Temperature of GISTEMP (Linear Regression)')
plt.xlabel('Years')
plt.ylabel('Means')
plt.show()
```

## Training the Polynomial Regression model on the whole dataset

# INDUSTRY1

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg1 = PolynomialFeatures(degree = 4)
X_poly = poly_reg1.fit_transform(x1)
lin_reg_1 = LinearRegression()
lin_reg_1.fit(X_poly, y1)
X_poly
```

# INDUSTRY2

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg2 = PolynomialFeatures(degree = 4)
X_poly = poly_reg2.fit_transform(x2)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y2)
X_poly
```

## Visualising the Polynomial Regression results

# INDUSTRY1

```python
plt.scatter(x1, y1, color = 'green')
plt.plot(x1, lin_reg_1.predict(poly_reg1.fit_transform(x1)),
         color = 'black')
plt.title('temperature of GCAG(Polynomial Regression)')
plt.xlabel('Years')
plt.ylabel('Means')
plt.show()
```

# INDUSTRY2

```python
plt.scatter(x2, y2, color = 'green')
plt.plot(x2, lin_reg_2.predict(poly_reg2.fit_transform(x2)),
         color = 'black')
plt.title('TEMPERATURE of GISTEMP (Polynomial Regression)')
plt.xlabel('Years')
plt.ylabel('Means')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

# INDUSTRY1

```python
X_grid = np.arange(min(x1), max(x1), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x1, y1, color = 'green')
plt.plot(X_grid, lin_reg_1.predict(poly_reg1.fit_transform(X_grid)), color = 'black')
plt.title('Temperature of GCAG (Polynomial Regression)')
plt.xlabel('Years')
plt.ylabel('Means')
plt.show()
```

# INDUSTRY2

```python
X_grid = np.arange(min(x2), max(x2), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x2, y2, color = 'green')
plt.plot(X_grid, lin_reg_2.predict(poly_reg2.fit_transform(X_grid)), color = 'black')
plt.title('Temperature of GISTEMP (Polynomial Regression)')
plt.xlabel('Years')
plt.ylabel('Means')
plt.show()
```

## Predicting a new result with Linear Regression

```python
lin_reg1.predict([[2017]])
```

```python
lin_reg2.predict([[2017]])
```

## Predicting a new result with Polynomial Regression

```python
lin_reg_1.predict(poly_reg1.fit_transform([[2017]]))
```

```python
lin_reg_2.predict(poly_reg2.fit_transform([[2017]]))
```
