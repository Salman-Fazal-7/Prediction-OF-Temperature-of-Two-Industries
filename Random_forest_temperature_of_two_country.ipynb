# Random Forest Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset= pd.read_csv('/content/annual_temp.csv')
dataset
```

# **Separate** **the** **data** **into** **two** **datasets**,**one** **for** **each** **country**

```python
Country1 = dataset[dataset["Source"].str.contains("GCAG")]
Country1
```

```python
x1 = Country1.iloc[:,1:2].values
y1 = Country1.iloc[:,-1].values
```

```python
x1
```

```python
y1
```

```python
Country2 = dataset[dataset["Source"].str.contains("GISTEMP")]
Country2
```

```python
x2 = Country2.iloc[:,1:2].values
y2 = Country2.iloc[:,-1].values
```

```python
x2
```

```python
y2
```

## Training the Random Forest Regression model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor1 = RandomForestRegressor(n_estimators = 100, random_state = 6)
regressor1.fit(x1, y1)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor2 = RandomForestRegressor(n_estimators = 100, random_state = 6)
regressor2.fit(x2, y2)
```

## Apply grids for higher resolution

```python
X_grid = np.arange(min(x1), max(x1), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
```

```python
X_grid = np.arange(min(x2), max(x2), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
```

## Visualizig

```python
plt.scatter(x1, y1, color = 'black')
plt.plot(X_grid, regressor1.predict(X_grid), color = 'green')
plt.title('Temperature of GCAG (Random Forest Regression)')
plt.xlabel('Year')
plt.ylabel('Means')
plt.show()
```

```python
plt.scatter(x1, y1, color = 'black')
plt.plot(X_grid, regressor1.predict(X_grid), color = 'green')
plt.title('Temperature of GISTEMP (Random Forest Regression)')
plt.xlabel('Year')
plt.ylabel('Means')
plt.show()
```

## Predicting a new result

```python
regressor1.predict([[2017]])
```
